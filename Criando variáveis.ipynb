```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv',sep=";")
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```

```python
dados_tipo_agregado = dados['Tipo Agregado']
dados_tipo_agregado.info()
```

```python
dados_tipo_agregado.describe()
```

```python
selecao = dados['Tipo Agregado'] == 'Casa'
dados[selecao].shape[0]
```

```python
dados[selecao].head(10)
```

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor Bruto','Valor m2']])
dados_aux.head(10)
```
