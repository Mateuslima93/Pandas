## Criando Agrupamentos

```python
import panda as pd
```

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
bairros = ['Barra da Tijuca','Copacabana','Ipanema','Leblon','Botafogo','Flamengo','Tijuca']
selecao = dados['Bairro'].isin(bairros)
selecao
```

```python
dados_bairro = dados[selecao]
dados_bairro
```

```python
grupo_bairros = dados.groupby('Bairro')
```

```python
for chave,valor in grupo_bairros:
    print(f'{chave} -> {valor.mean()}')
```

```python
grupo_bairros = dados_bairro.groupby('Bairro')
for chave, valor in grupo_bairros:
    print(f'{chave} -> {valor.Valor.mean()}')
```

```python
grupo_bairros[['Valor','Condominio']].mean().round(2)
```

## Estatísticas descritivas

```python
grupo_bairros['Valor'].aggregate(['min','max','sum'])
```
