## Seleções e frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv("dados/aluguel_residencial.csv", sep=";")
dados
```

```python
selecao = dados['Tipo'] == 'Apartamento'
selecao.head()
```

```python
dados_apartamento = dados.loc[selecao]
dados_apartamento
```

```python
dados_apartamento.describle()
```

```python
dados_apartamento.describe()
```

```python
selecao_casa = dados['Tipo'] == 'Casa'
selecao_casa_cond = dados['Tipo'] == 'Casa de Condominio'
selecao_casa_vila = dados['Tipo'] == 'Casa de Vila'
```

```python
dados_casas = dados.loc[(selecao_casa) | (selecao_casa_cond) | (selecao_casa_vila)]
dados_casas.head()
```

```python
dados_casas.describe()
```

```python
dados_casas.shape[0]
```

```python
selecao_area_menor = dados['Area'] >= 60
selecao_area_maior = dados['Area'] <= 100
```

```python
dados_area = dados.loc[selecao_area_menor & selecao_area_maior]
dados_area.describe()
```

```python
selecao_quartos = dados['Quartos'] >= 4
selecao_valor = dados['Valor'] < 2000
```

```python
dados_quartos_valor = dados.loc[selecao_quartos & selecao_valor]
dados_quartos_valor.describe()
```
