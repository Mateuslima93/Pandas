```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize =(15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head()
```

```python
area = plt.figure()
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
g1.scatter(dados['Valor'],dados['Area'])
g1.set_title('Valor X Area')
g2.hist(dados['Valor'])
g2.set_title('Distribuição Valor')
dados_g3 = dados['Valor'].sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)

grupo = dados.groupby('Tipo')['Valor']
g4.bar(grupo.mean().index,grupo.mean())
g4.set_title('Valores médios')
area
```

```python
area1 = plt.figure()
g1 = area1.add_subplot(1,2,1)
g2 = area1.add_subplot(1,2,2)

grupo_g1 = dados.groupby('Tipo Agregado')['Valor']
label_g1 = grupo_g1.count().index
valores_g1 = grupo_g1.count().values
g1.pie(valores_g1, labels=label_g1, autopct = '%1.1f%%')
g1.set_title('Tipo Agregado')

grupo_g2 = dados.groupby('Tipo')['Valor']
label_g2 = grupo_g2.count().index
valores_g2 = grupo_g2.count().values
g2.pie(valores_g2, labels=label_g2, autopct = '%1.1f%%')
g2.set_title('Tipos')
```
